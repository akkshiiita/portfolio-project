IMPORTING DEPENDENCIES

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to Pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
 #first five rows of dataset
credit_card_data.head()
```

```python
#last five rows of dataset
credit_card_data.tail()
```

```python
#dataset information
credit_card_data.info()
```

```python
#checking null values in dataset
credit_card_data.isnull().sum()
```

```python
#distribution of legit transaction and fraud transaction
credit_card_data['Class'].value_counts()
```

THIS DATASET IS HIGHLY UNBALANCED

1-> FRAUD
2-> LEGIT

```python
#separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measure for the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compares values for both transaction
credit_card_data.groupby('Class').mean()
```

under-sampling


building a sample dataset containing similar distribution of legit and fraud dataset

```python
legit_sample = legit.sample(n = 356)
```

concatenating two dataframes

```python
new_dataset = pd.concat([legit_sample,fraud],axis = 0)
#axis = 0 -> row, axis = 1 -> column
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features and targets

```python
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Spliting data into training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training -> Logistic Regression

```python
model = LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train,Y_train)
```

Model Evaluation

Accuracy score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('accuracy on training data : ', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy  = accuracy_score(X_test_prediction,Y_test)
```

```python
print('accuracy on test data : ', test_data_accuracy)
```
